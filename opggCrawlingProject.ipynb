```python
from bs4 import BeautifulSoup as bs
from urllib.request import urlopen
import re
import requests
import pandas as pd
```

# OP.GG 정보 크롤링하기

```python
url_opgg = 'https://www.op.gg/summoner/userName=%EB%A7%9B%EC%9E%88%EB%8A%94%EA%BD%81%EC%B9%98'

html = urlopen(url_opgg) # 서버에 결과 요청
soup = bs(html, "html.parser")

soup
```

```python
most_list.find('div', 'WinRatio').get_text()[6:9]
```

```python
def col_int(col):
  colInt_list = []

  for i in col:
    toInt = int(i.replace('\t', ''))
    toInt = int(i.replace('\n', ''))
    colInt_list.append(toInt)
  
  return colInt_list
```
