# **솔로 랭크 데이터 수집**

### \n,\t 제거 함수

```python
def col_int(col):
  colInt_list = []

  for i in col:
    toInt = i.replace('m', '')
    colInt_list.append(toInt)
  
  return colInt_list
```

### 분석 필요 데이터 리스트

```python
game_length = []
game_result = []
game_champion = []
game_kda = []
game_cs = []
```

```python
for i in range(5):
  g_length = soup.find_all('div', 'GameLength')[i].get_text()
  g_length = g_length.strip('\n')
  g_length = g_length.strip('\t')
  game_length.append(g_length)

  g_result = soup.find_all('div', 'GameResult')[i].get_text()
  g_result = g_result.strip('\n')
  g_result = g_result.strip('\t')
  game_result.append(g_result)

game_result
```
